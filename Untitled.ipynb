```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load and Preview the Data
df = pd.read_csv("customer_finance_data.csv")
df.head(5)
df.tail(5)
df.dtypes
df.shape
df[["Monthly_Income","Monthly_Expenses","Monthly_Savings"]].astype(int)
df.info()
```

```python
#Filtering and Indexing
print("Filter customers who are Married and whose Loan_Approved status is Yes.")
x = df[df["Marital_Status"].isin(["Married"])]
mar = x.loc[df["Loan_Approved"] == "Yes"]
print(mar[["Marital_Status","Loan_Approved"]])
print("Total count:",mar["Loan_Approved"].count())
print("---")
print("Use .iloc[] to display the Customer_ID, Monthly_Income, and Loan_Approved of the first 10 customers.")
print(df[["Customer_ID","Monthly_Income","Loan_Approved"]].iloc[1:10])
```

```python
#Descriptive Statistics
print("Calculate the average Monthly_Income.")
print(df["Monthly_Income"].mean())
print("---")
print("Use .describe() to generate summary statistics for all numeric columns.")
print(df[["Monthly_Income","Monthly_Expenses","Monthly_Savings"]].describe())
print("---")
print("Find the minimum and maximum value of Monthly_Savings.")
x = df.agg({
    "Monthly_Savings": ["min","max"]
})
print(x)
```

```python
#Value Counts and Sorting
print("Use .value_counts() to count occurrences of each Employment_Status.")
print(df["Employment_Status"].value_counts())
print("---")
print("Identify the most common Account_Type.")
print(df["Account_Type"].value_counts())
print("---")
print("Sort the entire dataset by Monthly_Income in descending order.")
ndf = df.sort_values(by="Monthly_Income", ascending=True)
ndf
```

```python
#Grouping and Aggregation
print("Group by Gender and find the average Monthly_Savings.")
print(df.groupby(["Gender"])["Monthly_Savings"].mean())
print("---")
print("Group by Marital_Status and count how many customers have Loan_Approved = Yes.")
la = df[df["Loan_Approved"].isin(["Yes"])]
print(la.groupby(["Marital_Status"])["Loan_Approved"].count())
print("---")
print("Group by both Gender and Employment_Status, then compute the average Monthly_Expenses.")
print(df.groupby(["Gender","Employment_Status"])["Monthly_Expenses"].mean())
```

```python
#Date-Based Insights
print("Convert the Join_Date column to datetime format.")
ndf["Join_Date"] = pd.to_datetime(ndf["Join_Date"])
ndf.info()
print("---")
print("How many customers joined in 2023?")
print("There are total of:",ndf["Join_Date"].dt.year.count())
print("---")
print("What is the average income of customers who joined after January 1, 2024?")
x = ndf[(ndf["Join_Date"].dt.year == 2024) & (ndf["Join_Date"].dt.month == 1) & (ndf["Join_Date"].dt.day == 1)]
print(x["Monthly_Income"].mean())
```

```python
#Boolean Filtering
print("Select customers with Monthly_Income > ₦300,000 and Monthly_Savings > ₦100,000.")
x = ndf[(ndf["Monthly_Income"] > 300000) & (ndf["Monthly_Income"] > 300000)]
x
```

```python
##Boolean Filtering
print("Use .isin() to find customers whose Employment_Status is either Unemployed or Self-employed")
x = df[df["Employment_Status"].isin(["Unemployed","Self-employed"])]
x
```

```python
#✅ Visualization Ideas Based on Your Exercise
print("Bar Chart")
am = ndf.groupby(["Gender"])["Monthly_Savings"].mean()
plt.figure(figsize=(5,3))
am.plot(kind="bar", title="Average Monthly Savings by Gender")
plt.ylabel("Monthly_Savings")
plt.xlabel("Gender")
plt.grid(True)
plt.show()
print("---")
print("Total Number of Customers by Employment Status")
es = ndf["Employment_Status"].value_counts()
plt.figure(figsize=(5,3))
es.plot(kind="bar", title="Total Number of Customers by Employment Status")
plt.ylabel("Number of Customers")
plt.xlabel("Employment Status")
plt.show()
print("---")
print("Average Monthly Expenses by Employment + Gender")
eeg = ndf.groupby(["Employment_Status","Gender"])["Monthly_Expenses"].mean()
plt.figure(figsize=(5,3))
eeg.plot(kind="bar", title="Average Monthly Expenses by Employment + Gender")
plt.show()
```

```python
#Pie Chart
print("Pie Chart")
print("Distribution of Account Types")
at = ndf["Account_Type"].value_counts()
plt.figure(figsize=(5,3))
at.plot(kind="pie", autopct='%1.1f%%', ylabel='Distribution of Account_Type')
plt.show()
print("---")
print("Proportion of Loan_Approved (“Yes” vs “No”)")
lp = ndf["Loan_Approved"].value_counts()
plt.figure(figsize=(5,3))
lp.plot(kind="pie", autopct='%1.1f%%', ylabel='Proportion of Loan_Approved')
plt.show()
```
